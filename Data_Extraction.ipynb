```python
#Authetication

auth = tweepy.OAuthHandler(api_key, api_key_secret)
auth.set_access_token(access_token, access_token_secret)
print("Authenticated")
```

```python
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
tweets_dataset= pd.DataFrame()
text = []
created_at =[]
tweet_id =[]
profile_background_image_url=[]
user_id =[]
user_name=[]
user_location=[]
user_description=[]

for i in range(0,100):
    
    text.append(tweets[i].text)
    created_at.append(tweets[i].created_at)
    tweet_id.append(tweets[i].id)
    user_id.append(tweets[i].user.id)
    user_name.append(tweets[i].user.name)
    user_location.append(tweets[i].user.location)
    user_description.append(tweets[i].user.description)
```

```python
tweets_dataset["text"]=text
tweets_dataset["created_at"]=created_at
tweets_dataset["tweet_id"]=tweet_id
tweets_dataset["user_id"]=user_id
tweets_dataset["user_name"]=user_name
tweets_dataset["user_location"]=user_location
tweets_dataset["user_description"]=user_description
```

```python
tweets_dataset.to_csv("Railway ministry mentioned tweets.csv")
```
