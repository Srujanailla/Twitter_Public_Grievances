# Separting @ from tweet text

```python
import re

tweets_dataset.text = [ re.sub("rt :",'',x)for x in tweets_dataset.text]
tweets_dataset.text = [ re.sub("\n",' ',x)for x in tweets_dataset.text]
tweets_dataset.text =  [re.sub(' +'," ", x) for x in tweets_dataset.text]
tweets_dataset.text =  [x.lower() for x in tweets_dataset.text]
tweets_dataset.text =  [re.sub("@[A-Za-z0-9_]+","", x) for x in tweets_dataset.text]
tweets_dataset.text =  [re.sub("#[A-Za-z0-9_]+","", x) for x in tweets_dataset.text]
tweets_dataset.text =  [re.sub(r"http\S+","", x) for x in tweets_dataset.text]
tweets_dataset.text =  [re.sub(r"www.\S+","", x) for x in tweets_dataset.text]
tweets_dataset.text =  [re.sub('[()!?]',"", x) for x in tweets_dataset.text]
tweets_dataset.text =  [re.sub('\[.*?\]',"", x) for x in tweets_dataset.text]
tweets_dataset.text =  [re.sub('\[.*?\]',"", x) for x in tweets_dataset.text]
```

```python
tweets_dataset.text =  [re.sub("[^A-Za-z0-9]"," ", x) for x in tweets_dataset.text]
```

```python
tweets_dataset.text =  [re.sub(' +'," ", x) for x in tweets_dataset.text]
```
